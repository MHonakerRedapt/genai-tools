```python
@tool
def get_links(query: str) -> str:
    """ Good for getting links to additional resources. """
    nearText = {"concepts": [query]}
    response = (
        weaviate_client.query
        .get("Sesame_128", ["links"])
        .with_near_text(nearText)
        .with_limit(5)
        .do()
        )
    response = ' , '.join([ ' , '.join(i['links']) for i in response['data']['Get']['Sesame_128']])
    return response

@tool
def get_information(query: str) -> str:
    """ Good for getting information for answering questions and chatting """
    nearText = {"concepts": [query]}
    response = (
        weaviate_client.query
        .get("Sesame_128", ["content"])
        .with_near_text(nearText)
        .with_limit(3)
        .do()
        )
    return ' '.join([text['content'] for text in response['data']['Get']['Sesame_128']])


@tool
def get_character_info(query: str) -> str:
    """
    Good for getting information about Sesame Street Characters.
    If the user question has one of the following names, use this tool.
    Abby Cadabby, Snuffleupagus, Big Bird, Baby Bear, Bert, Cookie Monster, 
    Count von Count, Ernie, Gabrielle, Grover, Gonger, Julia, Oscar the Grouch, 
    Rosita, Rudy, Tamir, Tango, Telly Monster, Zoe
    """

    response = (
        weaviate_client.query
        .get("Sesame_128", ["content"])
        .with_bm25(query=query, properties=['title'])
        .with_additional('score')
        .with_limit(2)
        .do()
        )

    return ' '.join([text['content'] for text in response['data']['Get']['Sesame_128']])

@tool
def get_character_links(query: str) -> str:
    """
    Good for getting information about Sesame Street Characters.
    If the user question has one of the following names, use this tool.
    Abby Cadabby, Snuffleupagus, Big Bird, Baby Bear, Bert, Cookie Monster, 
    Count von Count, Ernie, Gabrielle, Grover, Gonger, Julia, Oscar the Grouch, 
    Rosita, Rudy, Tamir, Tango, Telly Monster, Zoe
    """

    response = (
        weaviate_client.query
        .get("Sesame_128", ["links"])
        .with_bm25(query=query, properties=['title'])
        .with_additional('score')
        .with_limit(2)
        .do()
        )

    return ' '.join([text['links'] for text in response['data']['Get']['Sesame_128']])


tools = [get_information, get_links, get_character_info, get_character_links]
```

```python
def parallel_tasks(tasks):
    results = []
    with concurrent.futures.ThreadPoolExecutor() as exceutor:
        running_tasks = [exceutor.submit(task) for task in tasks]
        for running_task in running_tasks:
            results.append(running_task.result())
    return results
def x():
    return 2
def y():
    return 3
results = parallel_tasks([x, y])
#     lambda: print('IO task 1 running!'),
#     lambda: print('IO task 2 running!'),
# ])
results
```

```python
def run_io_tasks_in_parallel(tasks):
    with ThreadPoolExecutor() as executor:
        running_tasks = [executor.submit(task) for task in tasks]
        for running_task in running_tasks:
            running_task.result()

run_io_tasks_in_parallel([
    lambda: print('IO task 1 running!'),
    lambda: print('IO task 2 running!'),
])
```
